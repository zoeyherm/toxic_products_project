```python
from __future__ import print_function, division
import requests
requests.__path__
from bs4 import BeautifulSoup
from collections import defaultdict

import pickle
import random
import time

import pandas as pd
import numpy as np
```

```python
with open('all_shampoos_name_rating_link.pickle', 'rb') as read_file:
    df1 = pickle.load(read_file)
```

```python
df2 = df1[:300].copy()
```

```python
df2['1'] = 0
df2['2'] = 0
df2['3'] = 0
df2['4'] = 0
df2['5'] = 0
df2['6'] = 0
df2['7'] = 0
df2['8'] = 0
df2['9'] = 0
df2['10'] = 0

df2.head()
```

```python
for index, row in df2.iterrows():
    
    url = row['link']
    time.sleep(3 + 0.5*random.randint(1, 10))
    response = requests.get(url)
    print(response.status_code, index)
    
    page = response.text
    soup = BeautifulSoup(page, 'lxml')
    
    itemlist = []
    
    for item in soup.find_all(class_ = 'preview'):
        item = str(item)
        item = item.split('/')
        item = item[9].split('__')[0]
        item = item[-1]
        itemlist.append(str(item))  
    
    for item in itemlist:
        df2.at[index, item] += 1
```

```python
with open('0_to_299_cumul_ing.pickle', 'wb') as to_write:
    pickle.dump(df2, to_write)
```
